## Manim Scene

[Manim CLI](https://docs.manim.community/en/stable/guides/configuration.html)

**`%%manim -ql NewScene`** is a magic command.

> `%%manim` runs code cells with manim in a subprocess.<br>
`-ql` specifies low render quality (480p, 15fps).<br>
&emsp;&emsp;`-qm` — medium (720p, 30fps)<br>
&emsp;&emsp;`-qh` — high (1080p, 60fps)<br>
&emsp;&emsp;`-qk` — 4k (2160p, 60fps)<br>
`NewScene` is a scene class for rendering.

[Example Gallery](https://docs.manim.community/en/stable/examples.html)


```python
%%manim -qm ExampleScene

class ExampleScene(Scene):
    def construct(self):
        banner = ManimBanner()
        self.play(banner.create())
        self.play(banner.expand())
        self.wait(1.5)
```
