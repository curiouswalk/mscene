## Manim Scene

[Manim CLI](https://docs.manim.community/en/stable/guides/configuration.html)

**`%%manim -ql NewScene`** is a magic command.

> `%%manim` runs code cells with manim in a subprocess.<br>
`-ql` renders in 480p at 15fps.<br>
&emsp;&emsp;`-qm` — 720p, 30fps<br>
&emsp;&emsp;`-qh` — 1080p, 60fps<br>
&emsp;&emsp;`-qp` — 1440p, 60fps<br>
&emsp;&emsp;`-qk` — 2160p, 60fps<br>
`NewScene` is the scene class to be rendered.

[Manim Gallery](https://docs.manim.community/en/stable/examples.html)



```python
%%manim -qm ExampleScene

class ExampleScene(Scene):
    def construct(self):
        banner = ManimBanner()
        self.play(banner.create())
        self.play(banner.expand())
        self.wait(1.5)
```
